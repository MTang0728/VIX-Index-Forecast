### In this notebook, we will attempt to forecast the movment of VIX index using ARIMA model. 

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import VIX data file, check column names and dataframe size
vix_df = pd.read_csv('VIX.csv', parse_dates=['Date'])
print(vix_df.columns)
print('size of VIX data', vix_df.shape)
print('\n')
vix_df.head()
```

```python
# check if any columns contain missing values
vix_df.isnull().any()
```

```python
# set Date as the dataframe index
vix_df.set_index(['Date'], inplace= True)

# drop unused columns
drop_columns = ['Open', 'High', 'Low', 'Adj Close', 'Volume']
vix_df.drop(drop_columns, axis= 1, inplace= True)

# add a new column 'Return', which reflects percent change on price ('Close')
vix_df['Return'] = vix_df.Close.pct_change()
```

```python
vix_df.head()
```

```python
# create two new dataframes for analyzing Price and Return respectively
price_data = vix_df['Close']
return_data = vix_df['Return']
```

```python
%matplotlib inline
# plot Price data
plt.figure(figsize= (30, 10))
price_data.plot(kind = 'line')
plt.title('^VIX Daily Close Data')
plt.show()
```

```python
# Plot Return data
plt.figure(figsize= (30, 10))
return_data[:].plot(kind = 'line')
plt.title('^VIX Return Percentage Data')
plt.show()
```

```python
# split data into training and validation sets, using 0.75: 0.25
validation_ratio = 0.25
validation_size = int(len(return_data) * validation_ratio)

price_train, price_val = price_data[:-validation_size], price_data[-validation_size:]
return_train, return_val = return_data[:-validation_size], return_data[-validation_size:]

print('Training size is {}'.format(len(price_train)), '\nValidation size is {}'.format(len(price_val)))
```

### Check if the time series is stationary. Non-stationary time series tends to present index-dependent seasonality or trend. Statistically, the mean should stay relatively constant through the time series. If the time series is not stationary, differencing is required.
A simple way to check non-stationarity is to visually inspect the graph for any seasonal, trend effects that depend on tht eime index. A simple inspection of the graphs above shows that both time series are stationary. Another way to check for stationarity is by partitioning the time series and comparing the mean and variance of each group. Any significant differences would suggest that the original time series might be non-stationary

```python
# partition the data into two groups
partition_size = int(len(return_data) / 2)

price_train_1, price_train_2 = price_train[:partition_size], price_train[partition_size:]
return_train_1, return_train_2 = return_train[:partition_size], return_train[partition_size:]
# Check mean and variance of each partitions
print('Resutls on Price (Daily Close) Data:')
print('mean 1: {:.3f}, mean 2: {:.3f}'.format(price_train_1.mean(), price_train_2.mean()))
print('variance 1: {:.3f}, variance 2: {:.3f}'.format(price_train_1.var(), price_train_2.var()))
print('Results on Return Data:')   
print('mean 1: {:.3f}, mean 2: {:.3f}'.format(return_train_1.mean(), return_train_2.mean()))
print('variance 1: {:.3f}, variance 2: {:.3f}'.format(return_train_1.var(), return_train_2.var()))
```

```python
# use Augmented Dickey Fuller test to check if the time series is stationary
# the null hypothesis of ADF is that the time series is non-stationary, 
# if p-value is less than 0.05, the null hypothesis can be rejected
from statsmodels.tsa.stattools import adfuller

price_stationary = adfuller(price_train.dropna())
print('Results on Price (Daily Close) Data:')
print('ADF Statistics: {:.8f}'.format(price_stationary[0]))
print('p-value: {:.8f}'.format(price_stationary[1]))

return_stationary = adfuller(return_train.dropna())
print('Results on Return Data:')
print('ADF Statistics: {:.8f}'.format(return_stationary[0]))
print('p-value: {:.8f}'.format(return_stationary[1]))
```

```python
# use pmdarima built-in function to assess stationarity
from pmdarima.arima import ndiffs

price_kpss_diffs = ndiffs(price_train, alpha=0.05, test='kpss', max_d=6)
price_adf_diffs = ndiffs(price_train, alpha=0.05, test='adf', max_d=6)
price_n_diffs = max(price_adf_diffs, price_kpss_diffs)

return_kpss_diffs = ndiffs(return_train.dropna(), alpha=0.05, test='kpss', max_d=6)
return_adf_diffs = ndiffs(return_train.dropna(), alpha=0.05, test='adf', max_d=6)
return_n_diffs = max(return_adf_diffs, return_kpss_diffs)

print('Results on Price (Daily Close) Data:')
print(f"Estimated differencing term: {price_n_diffs}")
print('Results on Return Data:')
print(f"Estimated differencing term: {return_n_diffs}")
```

```python
# import the libraries needed to plot autocorrelation and partial autocorrelation
from pandas.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# plot autocorrelation of daily close price
plt.rcParams.update({'figure.figsize':(20,7)})
plot_pacf(price_train.dropna(), lags = 100);
```

```python
# plot partial autocorrelation of daily close price
plt.rcParams.update({'figure.figsize':(20, 10)})
plot_acf(price_train.dropna(), lags = 300);
```

```python
# import AutoARIMA module
import pmdarima as pm
# train an auto arima model, set d to be 1 as obtained from previous steps
price_model = pm.auto_arima(price_train.dropna(), d = 1, 
                      seasonal= False, stepwise= True, 
                      suppress_warnings= True, 
                      error_action= 'ignore', 
                      max_p = 6, max_q = 6, 
                      max_order= None, 
                      trace = True)
```

```python
# optimal order generated by autoARIMA model
print(price_model.order)
```

```python
from statsmodels.tsa.arima_model import ARIMA

# using order (1,1,1), build an ARIMA Model usinfg a difference ARIMA package and print model summary
arima_price_model = ARIMA(price_train, order=(1,1,1))
arima_price_model_fit = arima_price_model.fit(disp=0)
print(arima_price_model_fit.summary())
```

```python
# import packages for evaluating forecast
from sklearn.metrics import mean_squared_error
from pmdarima.metrics import smape
from sklearn.metrics import r2_score
```

```python
# define a funciton for making predictions
def forecast_one_step(model):
    fc, conf_int = model.predict(n_periods= 1, return_conf_int= True)
    return (fc.tolist()[0], np.asarray(conf_int).tolist()[0])
# define lists for storing forecasted values and their confidence intervals
price_forecasts = []
price_confidence_intervals = []
# define a counter variable for tracking model forecasting progress
check_point = 0
# loop through validation data and make predictions
for new_ob in price_val:
    price_fc, price_conf = forecast_one_step(price_model)
    price_forecasts.append(price_fc)
    price_confidence_intervals.append(price_conf)
    # update ARIMA model by adding new observation to the model
    price_model.update(new_ob)
    # print out progress. 190 is used here because its close to 1/10 of the size of the validation data
    if (len(price_forecasts) % 190 == 0):
        check_point += 1
        print('Progress: {}%'.format(check_point * 10))
```

```python
# print evaluation metrics
print("Mean squared error: {}".format(mean_squared_error(price_val, price_forecasts)))
print("SMAPE: {}".format(smape(price_val, price_forecasts)))
print("R^2 score: {}".format(r2_score(price_val, price_forecasts)))
```

```python
# plot forecasted results vs actual results
fig, axes = plt.subplots(2, 1, figsize=(20, 15))
# plot both training and forecasted data
price_conf_int = np.asarray(price_confidence_intervals)
axes[0].plot(price_train, color='blue', label='Training Data')
axes[0].plot(price_val.index, price_forecasts, color='green', label='Predicted Price')
axes[0].plot(price_val.index, price_val, color='red', label='Actual Price')
axes[0].fill_between(price_val.index, 
                      price_conf_int[:,0], price_conf_int[:, 1],
                      alpha = 0.5, color = 'orange',
                      label = 'Confidence Intervals')
axes[0].legend()
# plot forecasted data only
axes[1].plot(price_val[-240:].index, price_forecasts[-240:], color='green', label='Predicted Price')
axes[1].plot(price_val[-240:].index, price_val[-240:], color='red', label='Actual Price')
axes[1].fill_between(price_val[-240:].index, 
                      price_conf_int[-240:,0], price_conf_int[-240:, 1],
                      alpha = 0.5, color = 'orange',
                      label = 'Confidence Intervals')
axes[1].legend()
```
